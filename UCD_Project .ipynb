```python
rock_tracks_data = pd.read_csv('rock_playlist_tracks_data.csv')
rock_tracks_data_new = rock_tracks_data.drop_duplicates(subset=['track_name'], keep='first')
rock_tracks_data_new = rock_tracks_data.replace(0, np.nan)
rock_tracks_data_new = rock_tracks_data.drop(['key', 'mode', 'valence'], axis=1)
rock_tracks_data_new.head()
```

```python
rock_tracks_data_new.dtypes
```

```python
new_column_years = rock_tracks_data_new['release_date']
new_column_years
```

```python
def years(year):
        if len(year) == 4:
            return year
        else:
            return year[0:4]
rock_tracks_data_new['year'] = rock_tracks_data_new['release_date'].apply(years)
rock_tracks_data_new.head()
```

```python
energy = rock_tracks_data_new[rock_tracks_data_new.popularity != 0]
sns.scatterplot(data=energy, x='energy', y='popularity')
plt.title("Correlation between Popularity & Energy", fontsize=15)
plt.xlabel("Energy", fontsize=12)
plt.ylabel("Popularity",fontsize=12)
plt.show()
```

```python
sns.pairplot(data=rock_tracks_data_new, vars=['popularity', 'loudness', 'energy', 'tempo', 'liveness', 'danceability'],
             diag_kws={'color':'red'}, plot_kws={'color':'green'})
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(rock_tracks_data_new.corr(), cmap='RdBu', vmin=-1, vmax=1, annot=True)
```
